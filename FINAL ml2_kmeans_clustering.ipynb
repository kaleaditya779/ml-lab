implement k-means clustering and determine optimal number of clusters using elbow method:

1) Determine the optimal number of clusters using elbow method.
2) Do k-means clustering on the data using the calculated optimal cluster count.

```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sb
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('sales_data_sample.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
features = df[['QUANTITYORDERED', 'PRICEEACH', 'SALES']]
```

```python
scalar = StandardScaler()
scaled_features = scalar.fit_transform(features)
```

```python
# Elbow method to determine optimal number of clusters:
inertia = []
# Inertia is the sum of squared distances to the closest centroid

for k in range(1, 10):
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    kmeans_model.fit(scaled_features)
    inertia.append(kmeans_model.inertia_)
```

```python
plt.plot(range(1,10), inertia, marker='o')
plt.xlabel('no of clusters')
plt.ylabel('sum of squared distances')
```

```python
# elbow point = 3 (optimal cluster count)
kmeans_model = KMeans(n_clusters=3, random_state=42)
kmeans_model.fit(scaled_features)
```

```python
labels = kmeans_model.predict(scaled_features)
df['cluster'] = labels
```

```python
df.head()
```

Evaluating the clustering:

```python
sill_score = silhouette_score(X=scaled_features, labels=labels)
```

```python
sill_score
```

```python
# Display cluster centers
print("Cluster Centers:\n", scalar.inverse_transform(kmeans_model.cluster_centers_))
```
